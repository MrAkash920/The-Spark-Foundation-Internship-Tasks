### **Supervised Learning**
Supervised learning is the types of machine learning in which machines are trained using well "labelled" training data, and on basis of that data, machines predict the output. The labelled data means some input data is already tagged with the correct output.

### **Linear Regression**
Linear regression algorithm shows a linear relationship between a dependent (y) and one or more independent (y) variables, hence called as linear regression. Since linear regression shows the linear relationship, which means it finds how the value of the dependent variable is changing according to the value of the independent variable.


## Tareget
Predict the percentage of an student based on the number of study hours


```python
# Import required libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
%matplotlib inline
```

```python
# import data from url
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df.head(10) #print top 10 data.
```

##  Data Visulizatioin

```python
# Visualization of the realtionship between both variables using pair plot
sns.pairplot(df)
plt.title('Pairplot of Hours and Scores')
plt.show()
```

```python
# Create a scatterplot to visualize the relationship between Hours and Scores
sns.scatterplot(data=df, x='Hours', y='Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Scatter Plot of Hours vs. Scores')
plt.show()
```

```python
# Create a distribution plot for Hours
sns.histplot(data=df, x='Hours', kde=True, color='skyblue', bins=10)
plt.xlabel('Hours')
plt.title('Distribution of Hours')
plt.show()
```

Prepare the data

```python
X = df[['Hours']]
y = df['Scores']
```

Split the data into a training set and a testing set (80% training, 20% testing)


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Create a Linear Regression model and fit it to the training data


```python
model = LinearRegression()
model.fit(X_train, y_train)
```

Make predictions using the test set


```python
y_pred = model.predict(X_test)
```

Visualize the regression line

```python
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Regression Line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

Evaluate the model's performance


```python
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared (R2) Score: {r2}')
```

Predict the score for a student who studies 9.25 hours/day


```python
hours = 9.25
predicted_score = model.predict([[hours]])
print(f'Predicted Score for {hours} hours of study: {predicted_score[0]}')
```
